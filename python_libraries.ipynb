##### Creating a pandas framework

```python
import pandas as pd

data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Kenneth', 'Sandra'],
    'Age': [24, 27, 22, 35, np.nan],
    'City': ['Edo', 'Ibadan', 'Warri', 'Edo', np.nan]
}
df = pd.DataFrame(data_dict)
df
```

```python
print(type(pd.DataFrame))
```

###### filter database on certain conditions

```python
filtered_df = df[(df['Age'] > 30) & (df['City'] == 'Edo')]
```

```python
df['Age'].sum()
```

```python
df['Age'].mean()
```

```python
df['Math score'] = [70, 32, 58, 81, 60]
df
```

```python
df['English score'] = [85, 81, 73, 92, 70]
df['Score'] = (df['Math score'] + df['English score']) / 2
df
```

### Data Cleaning

```python
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
df
```

```python
fill_city = 'Lagos'
df['City'] = df['City'].fillna(fill_city)
df
```
