#### Write a python program to count the number of unique values in the list below <br /> sample_list = [20, 90, 21, 20, 39, 89, 90, 21, 20, 39, 89, 100, 2, 1, 2, 89, 90, 21, 1]

```python
sample_list = [20, 90, 21, 20, 39, 89, 90, 21, 20, 39, 89, 100, 2, 1, 2, 89, 90, 21, 1]
s = set()
for c in sample_list:
    s.add(c)
s
len(s)
```
